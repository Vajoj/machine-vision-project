```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image
img = cv2.imread("Training_Data/4K7_1-4W/4K7_1-4W_(3).jpg")  # Replace with your actual image path
### **Step 1: Preprocessing**
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
smooth = cv2.bilateralFilter(gray, 5, 150, 150)
plt.imshow(smooth, cmap='gray')
plt.title("Blurred")
plt.show()

# background = cv2.imread("Training_Data/wood2.jpg", cv2.IMREAD_GRAYSCALE)
# background_resized = cv2.resize(background, (smooth.shape[1], smooth.shape[0]))
# plt.imshow(background_resized, cmap='gray')
# plt.title("background")
# plt.show()
# diff = smooth#cv2.absdiff(smooth, background_resized) 

# plt.imshow(diff, cmap='gray')
# plt.title("diff")
# plt.show()

# # Apply histogram equalization
# equalized_img = cv2.equalizeHist(clahe_img)

# # Display the original and equalized images
# plt.figure(figsize=(10, 5))

# plt.subplot(1, 2, 1)
# plt.imshow(smooth, cmap='gray')
# plt.title("Blurred")
# plt.axis('off')

# plt.subplot(1, 2, 2)
# plt.imshow(equalized_img, cmap='gray')
# plt.title("Equalized Image")
# plt.axis('off')

# plt.show()

# Thresholding to enhance object isolation
adaptive_binary = cv2.adaptiveThreshold(smooth, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 51, 10)


plt.imshow(adaptive_binary, cmap='gray')
plt.title("Threshold")
plt.show()

adaptive_binary = cv2.bilateralFilter(adaptive_binary, 5, 150, 150)
plt.imshow(smooth, cmap='gray')

# Edge detection
edges = cv2.Canny(adaptive_binary, 170, 200)

plt.imshow(edges, cmap='gray')
plt.title("Edges")
plt.show()

# Morphological operations for better shape closure
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (13, 13))
processed = cv2.morphologyEx(edges, cv2.MORPH_CLOSE, kernel, iterations=3)
processed = cv2.morphologyEx(processed, cv2.MORPH_OPEN, kernel, iterations=1)

plt.imshow(processed, cmap='gray')
plt.title("Processed Image Before Contour Detection")
plt.show()

# Find contours
contours, _ = cv2.findContours(processed, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

print(f"Contours detected before filtering: {len(contours)}")

# Debugging: Visualize contours (Main Output)
debug_contours = img.copy()
cv2.drawContours(debug_contours, contours, -1, (255, 0, 0), 2)

# Display debug visualization as the main result
plt.imshow(cv2.cvtColor(debug_contours, cv2.COLOR_BGR2RGB))
plt.title("Final Contours Debug Output")
plt.axis("off")
plt.show()

# Filter contours based on shape properties
filtered_contours = []
for cnt in contours:
    area = cv2.contourArea(cnt)
    if area > 100:  # Adjust area threshold based on your image
        filtered_contours.append(cnt)

# Ensure at least one contour is found
if not filtered_contours:
    print("Error: No valid resistor contour detected.")
    exit()

# Find bounding box for the largest contour
x, y, w, h = cv2.boundingRect(filtered_contours[0])

# Use cv2.minAreaRect to get the angle of the resistor
rect = cv2.minAreaRect(filtered_contours[0])
angle = rect[2]

# To ensure horizontal orientation, adjust the angle to match the horizontal axis
if angle < -45:
    angle = 90 + angle

# Rotate the image to align the resistor horizontally
center = (x + w // 2, y + h // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
rotated_img = cv2.warpAffine(img, rotation_matrix, (img.shape[1], img.shape[0]))

# Reapply the same preprocessing steps to the rotated image
gray_rotated = cv2.cvtColor(rotated_img, cv2.COLOR_BGR2GRAY)
smooth_rotated = cv2.bilateralFilter(gray_rotated, 5, 150, 150)
adaptive_binary_rotated = cv2.adaptiveThreshold(smooth_rotated, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 51, 10)
edges_rotated = cv2.Canny(adaptive_binary_rotated, 170, 200)

# Morphological operations on the rotated image
processed_rotated = cv2.morphologyEx(edges_rotated, cv2.MORPH_CLOSE, kernel, iterations=3)
processed_rotated = cv2.morphologyEx(processed_rotated, cv2.MORPH_OPEN, kernel, iterations=1)

# Find contours in the rotated image
contours_rotated, _ = cv2.findContours(processed_rotated, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Filter contours for the rotated image
filtered_contours_rotated = []
for cnt in contours_rotated:
    area = cv2.contourArea(cnt)
    if area > 100:  # Adjust area threshold based on your image
        filtered_contours_rotated.append(cnt)

# Ensure at least one valid contour is found after rotation
if not filtered_contours_rotated:
    print("Error: No valid resistor contour detected in rotated image.")
    exit()

# Find bounding box for the largest contour in the rotated image
x_rot, y_rot, w_rot, h_rot = cv2.boundingRect(filtered_contours_rotated[0])

# Crop the rotated resistor
cropped_resistor_rotated = rotated_img[y_rot:y_rot + h_rot, x_rot:x_rot + w_rot]

# Show the rotated and cropped resistor
plt.imshow(cv2.cvtColor(cropped_resistor_rotated, cv2.COLOR_BGR2RGB))
plt.title("Rotated and Cropped Resistor")
plt.axis("off")
plt.show()

# Create a blank mask
mask = np.zeros_like(gray)

# Fill the contour on the mask (keeping only the resistor body)
cv2.drawContours(mask, filtered_contours, -1, 255, thickness=cv2.FILLED)

# Apply the mask to the original image
extracted_resistor = cv2.bitwise_and(img, img, mask=mask)

# Find bounding box and crop the resistor
x, y, w, h = cv2.boundingRect(filtered_contours[0])  # Assuming the largest contour is the resistor
cropped_resistor = extracted_resistor[y:y+h, x:x+w]  # Crop only the resistor region

# Show the isolated resistor
plt.imshow(cv2.cvtColor(cropped_resistor, cv2.COLOR_BGR2RGB))
plt.title("Extracted Resistor for Color Analysis")
plt.axis("off")
plt.show()

if len(contours) == 1:
    cnt = contours[0]  # Get the single contour

    # Existing code for orientation check and flipping
    x, y, w, h = cv2.boundingRect(cnt)
    aspect_ratio = h / float(w) if w != 0 else 0  # Avoid division by zero

    # Check if the aspect ratio indicates verticality
    if aspect_ratio > 1.5:  # Vertical orientation
        print("Vertical resistor detected, flipping debug output...")

        # Flip the debug output vertically
        cropped_resistor = cv2.rotate(cropped_resistor, cv2.ROTATE_90_CLOCKWISE)

        # Display the flipped debug output
        plt.imshow(cv2.cvtColor(cropped_resistor, cv2.COLOR_BGR2RGB))
        plt.title("Flipped Debug Output")
        plt.axis("off")
        plt.show()
    else:
        print("Horizontal resistor detected; no flip applied.")
else:
    print("Error: More than one contour detected, expected one resistor.")


# SHARPEN
sharpen_kernel = np.array([
    [0, -1, 0],
    [-1, 5,-1],
    [0, -1, 0]
])
sharpened = cv2.filter2D(cropped_resistor, -1, sharpen_kernel)

# CLAHE for vibrant colors
hsv_sharp = cv2.cvtColor(sharpened, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv_sharp)

clahe = cv2.createCLAHE(clipLimit=1.0, tileGridSize=(3,3))
v_clahe = clahe.apply(v)

enhanced_hsv = cv2.merge((h, s, v_clahe))
vibrant_resistor = cv2.cvtColor(enhanced_hsv, cv2.COLOR_HSV2BGR)

# denoising
vibrant_resistor = cv2.bilateralFilter(vibrant_resistor, 9, 75, 75)

### **Step 3: Color Classification for Resistor Bands**

plt.imshow(cv2.cvtColor(vibrant_resistor, cv2.COLOR_BGR2RGB))
plt.title("Original BGR Image")
plt.axis("off")
plt.show()

# Convert to HSV for color segmentation
img_hsv = cv2.cvtColor(vibrant_resistor, cv2.COLOR_BGR2HSV)
plt.imshow(cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB))
plt.title("HSV Image")
plt.axis("off")
plt.show()

h, s, v = cv2.split(img_hsv)

h_normalized = (h * 255 / 179).astype(np.uint8)

fig, axes = plt.subplots(1, 3, figsize=(20, 5))
axes[0].imshow(h, cmap="jet"), axes[0].set_title("Hue Channel"), axes[0].axis("off")
axes[1].imshow(s, cmap="gray"), axes[1].set_title("Saturation Channel"), axes[1].axis("off")
axes[2].imshow(v, cmap="gray"), axes[2].set_title("Value Channel"), axes[2].axis("off")

plt.show()

# Define HSV color ranges for resistor bands
color_ranges = {
    'black': [(0, 0, 5), (180, 255, 50)],
    'brown': [(7, 100, 50), (17, 255, 120)],
    'red': [(0, 160, 100), (4, 255, 255)],
    'orange': [(4, 160, 100), (20, 255, 255)],
    'yellow': [(20, 100, 100), (35, 255, 255)],
    'green': [(35, 100, 40), (85, 255, 255)],
    'blue': [(85, 100, 50), (130, 255, 255)],
    'violet': [(130, 50, 50), (160, 255, 255)],
    'grey': [(0, 0, 75), (180, 50, 200)],
    'white': [(0, 0, 200), (180, 30, 255)],
    'gold': [(14, 100, 120), (32, 220, 220)],
    'silver': [(0, 0, 100), (180, 15, 200)],
}

height, width = img_hsv.shape[:2]
# Create a binary mask with 1s in the middle, 0s on borders
resistor_mask = np.zeros((height, width), dtype=np.uint8)
cv2.rectangle(resistor_mask, (int(0.1*width), int(0.1*height)), (int(0.9*width), int(0.9*height)), 255, -1)
plt.imshow(resistor_mask, cmap="gray")
plt.title(f"Mask to remove edges")
plt.axis("off")
plt.show()

kernel = np.ones((3, 3), np.uint8)

def calculate_confidence(masked_region, hsv_image, color_range):
    # Count total pixels in the masked region
    total_pixels = np.sum(masked_region > 0)

    # Create mask based on the given color range
    lower_bound, upper_bound = np.array(color_range[0]), np.array(color_range[1])
    color_mask = cv2.inRange(hsv_image, lower_bound, upper_bound)

    # Count pixels within the color range
    matched_pixels = np.sum(np.bitwise_and(masked_region, color_mask) > 0)

    # Calculate confidence score (percentage of matched pixels)
    confidence_score = (matched_pixels / total_pixels) * 100 if total_pixels > 0 else 0
    return confidence_score

for color, (lower, upper) in color_ranges.items():
    # Create a mask for the color
    mask = cv2.inRange(img_hsv, np.array(lower), np.array(upper))

    # Apply the resistor center mask to ignore edges
    masked = cv2.bitwise_and(mask, resistor_mask)

    # Apply morphological operations to clean up the mask
    masked_cleaned = cv2.morphologyEx(masked, cv2.MORPH_CLOSE, kernel)  # Dilation followed by erosion
    masked_cleaned = cv2.morphologyEx(masked_cleaned, cv2.MORPH_OPEN, kernel)  # Erosion followed by dilation

    # Show the result of the final masked region
    plt.imshow(masked_cleaned, cmap="gray")
    plt.title(f"Cleaned Mask for {color} (center only)")
    plt.axis("off")
    plt.show()

    # Use the cleaned masked version for contour detection
    contours, _ = cv2.findContours(masked_cleaned, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    filtered_contours = []

    for cnt in contours:
        x, y, w, h = cv2.boundingRect(cnt)
        aspect_ratio = h / float(w) if w != 0 else 0

        area = cv2.contourArea(cnt)

        # Check if the contour is valid
        if area > 15 and aspect_ratio > 1.2 and h > 5 and w > 0.5:
            filtered_contours.append(cnt)

            # Confidence check and drawing
            confidence = calculate_confidence(masked_cleaned, img_hsv, [lower, upper])
            if confidence > 70:  # Only print and draw if the confidence is above threshold
                print(f"Accepted {color} band with confidence: {confidence:.2f}% at x={x}, area={area:.1f}, AR={aspect_ratio:.2f}")
                
                # Draw the contour on the preview image
                preview = vibrant_resistor.copy()  # Copy the original image for visualization
                cv2.drawContours(preview, [cnt], -1, (0, 255, 0), 2)

                plt.imshow(cv2.cvtColor(preview, cv2.COLOR_BGR2RGB))
                plt.title(f"Contours for {color} (Filtered and Accepted)")
                plt.axis("off")
                plt.show()
            else:
                print(f"Rejected {color} band due to low confidence: {confidence:.2f}%")

    # Optionally, display all detected bands in a summary plot
    print(f"Filtered vertical band contours for {color}: {len(filtered_contours)}")


band_info = []

for color, (lower, upper) in color_ranges.items():
    mask = cv2.inRange(img_hsv, np.array(lower), np.array(upper))
    masked = cv2.bitwise_and(mask, resistor_mask)
    masked_cleaned = cv2.morphologyEx(masked, cv2.MORPH_CLOSE, kernel)
    masked_cleaned = cv2.morphologyEx(masked_cleaned, cv2.MORPH_OPEN, kernel)

    contours, _ = cv2.findContours(masked_cleaned, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area < 20:
            continue  # Skip tiny specks

        x, y, w, h = cv2.boundingRect(cnt)
        cx = x + w // 2
        aspect_ratio = h / float(w) if w != 0 else 0

        if 1.0 < aspect_ratio < 10:  # More inclusive ratio
            band_info.append((cx, color))
            cv2.rectangle(cropped_resistor, (x, y), (x + w, y + h), (0, 255, 0), 1)
            print(f"Added {color} band at x={cx}, area={area}, AR={aspect_ratio:.2f}")



print("Detected band colors:", band_colors)

color_digit = {
    'black': 0, 'brown': 1, 'red': 2, 'orange': 3,
    'yellow': 4, 'green': 5, 'blue': 6, 'violet': 7,
    'grey': 8, 'white': 9
}

multiplier_map = {
    'black': 1, 'brown': 10, 'red': 100, 'orange': 1_000,
    'yellow': 10_000, 'green': 100_000, 'blue': 1_000_000,
    'gold': 0.1, 'silver': 0.01
}

def calculate_resistor_value(bands):
    if len(bands) == 4:
        digit1 = color_digit.get(bands[0], 0)
        digit2 = color_digit.get(bands[1], 0)
        multiplier = multiplier_map.get(bands[2], 1)
        return (10 * digit1 + digit2) * multiplier
    elif len(bands) == 5:
        digit1 = color_digit.get(bands[0], 0)
        digit2 = color_digit.get(bands[1], 0)
        digit3 = color_digit.get(bands[2], 0)
        multiplier = multiplier_map.get(bands[3], 1)
        return (100 * digit1 + 10 * digit2 + digit3) * multiplier
    else:
        return None

# Sort left-to-right
band_info.sort(key=lambda x: x[0])
band_colors = [color for _, color in band_info]

print("Sorted band colors:", band_colors)

# Flip if tolerance color is first
tolerance_colors = ['gold', 'silver']
if band_colors and band_colors[0] in tolerance_colors:
    print("Tolerance band appears first, flipping band order!")
    band_colors.reverse()

# Sanity check
if len(band_colors) == 4:
    if all(b in color_digit for b in band_colors[:2]) and band_colors[2] in multiplier_map:
        value = calculate_resistor_value(band_colors)
        print(f"Resistor value: {value} ohms")
    else:
        print("Invalid band sequence for 4-band resistor.")
else:
    print("Unsupported number of bands.")
```
